# Atividade 0 : Transformação Linear

Importei a biblioteca numpy numpy para realizar calculos de matrizes e matplotlib para ver os gráficos.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def aplicar_transformacao(vetor, matriz):
    return vetor @ matriz
```

## Os Vetores unitários são: 

$$
\hat i=\begin{bmatrix}
1 \\
0
\end{bmatrix}

,\quad\hat j=\begin{bmatrix}
0 \\
1
\end{bmatrix}
$$

Gerando a Matriz : 
$$
\begin{bmatrix}
1&0\\
0&1
\end{bmatrix}


$$

```python
plt.quiver(0, 0, 1, 0, angles='xy', scale_units='xy', scale=1, color='blue')
plt.quiver(0, 0, 0, 1, angles='xy', scale_units='xy', scale=1, color='red')
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.grid(True)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
```

## Aplicando rotação de 90 graus no sentido horário
Os vetores unitários após a rotação:

$$
\hat i=\begin{bmatrix}
0 \\
-1
\end{bmatrix}

,\quad\hat j=\begin{bmatrix}
1 \\
0
\end{bmatrix}
$$

Gerando a Matriz : 
$$
A=\begin{bmatrix}
0&1\\
-1&0
\end{bmatrix}


$$


```python
matrizA = np.array([[0, 1], [-1, 0]])
```

```python
plt.quiver(0, 0, 0, -1, angles='xy', scale_units='xy', scale=1, color='blue')
plt.quiver(0, 0, 1, 0, angles='xy', scale_units='xy', scale=1, color='red')
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.grid(True)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.show()

```

## Exemplo com Vetores arbitrários

$$
\hat v_1=\begin{bmatrix}
7 \\
7
\end{bmatrix}

,\quad\hat v_2=\begin{bmatrix}
3 \\
8
\end{bmatrix}
$$

$$
Matriz\quad dos\quad vetores\quad arbitrários=\begin{bmatrix}
7&3 \\
7&8
\end{bmatrix}
$$



```python
matrizB = np.array([[7, 3], [7, 8]])
```

```python
vetoresRotacionados = aplicar_transformacao(matrizA, matrizB)
print(vetoresRotacionados)
```

```python
plt.quiver(0, 0, 7, 7, angles='xy', scale_units='xy', scale=1, color='blue')
plt.quiver(0, 0, 3, 8, angles='xy', scale_units='xy', scale=1, color='red')
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.xticks(range(0, 11))
plt.yticks(range(0, 11))
plt.grid(True)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.show()
```

## Aplicando Shear em X

Irei fazer o o Shearmento horizontal usando 1 como uma constante de proporcionalidade
Para $$ \hat v_1: x = 7 + 1.(-7) = 0,\quad y = -7$$
$$ \hat v_2: x = 8 + 1.(-3) = 5,\quad y = -3$$
### Resultando em:
$$
\hat v_1=\begin{bmatrix}
0 \\
-7
\end{bmatrix}
,\quad\hat v_2=\begin{bmatrix}
5 \\
-3
\end{bmatrix}
$$
$$
Matriz\quad dos\quad vetores\quad arbitrários \quad rotacionados \quad após \quad shearmento=\begin{bmatrix}
0&5 \\
-7&-3
\end{bmatrix}
$$

```python
matrizShear = (np.array([[1, 1], [0, 1]]))
matrizShearda = aplicar_transformacao(matrizShear, vetoresRotacionados)
print(matrizShearda)
```

```python
plt.quiver(0, 0, matrizShearda[0,0], matrizShearda[1,0], angles='xy', scale_units='xy', scale=1, color='blue')
plt.quiver(0, 0, matrizShearda[0,1], matrizShearda[1,1], angles='xy', scale_units='xy', scale=1, color='red')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.xticks(range(-10, 11))
plt.yticks(range(-10, 11))
plt.grid(True)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.show()
```
